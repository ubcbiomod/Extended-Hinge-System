```python
import json


def sandwich(inputvar):
    
    ####################################################################################################################
    
    lisofscaf = []
    lisofscaflen = []
    lisofstap = []
    lisofstaplen = []
    lisofind = []
    lisofstapind = []
    lisofscafindL = []
    lisofscafindR = []
    lisofstapL = []
    lisofstapR = []
    lisofstapindL = []
    lisofstapindR = []
    lisofscafLnum = []
    lisofscafRnum = []
    lisofstapLnum = []
    lisofstapRnum = []
    DNAsort = []
    DNAoverall = []
    Sandwichsort = []
    Sandwichoverall = []
    
    # This section of the code defines the starting lists of the system. Most of these will be empty, with values
    # appended to them later down the line.
    
    # Note, functionality for lisofscaf/scaffold strand related variables currently have no function in this code,
    # as it has been taken out.
    
    # lisofscaf is the list of scaffolds in the .json
    
    # lisofscaflen is a list of the lengths of the scaffolds
    
    # lisofstap is the list of staple strands in the .json
    
    # lisofstaplen is the list of lengths of staple strands
    
    # lisofind is the numerical position of the honey comb, relative to 0
    
    # lisofstapind serves the same function as above, but is differentiated to separate variables for function
    
    # lisof___L are the horizontal coordinates of anytime a cross over appears on the left side for a single item
    
    # lisof___R are the horizontal coordinates of anytime a cross over appears on the right side for a single item
    
    # lisof___ind_cross will be used to determine lengths of base strands in the specified crossover
    
    # lisof___R or L are the horizontal coordinates of all occurances of crossovers/squares/arrows that occur in the entire list.
    
    # DNAsort is the list that contains an ith sorted single strand scaffold in the order of initialization, crossover
    # and termination.
    
    # DNAoverall is the nested list of all DNA scaffold strands locations i.e. a list of DNAsort
    
    # Sandwichsort is similar as DNAsort, except the functionality is meant for staple strands
    
    # Sandwichoverall serves the same purpose as DNAoverall
    
    ####################################################################################################################
    
    fh = open(inputvar,'r')
    line = fh.readline()
    inp = json.loads(line)
    lis = inp['vstrands']
    
    # The section above will read some string "x" representing a .json file from cadnano. This file will be read into
    # a list called lis.
    
    # SECTION 1 START: This entire section checks for where our location of crossovers are at (Modularity easy)
    
    ####### SUBSECTION 1: Index the string within lis to obtain either scaf, stap, or num data.   #########################
    
    for i in range(len(lis)):
        lisofscaf.append(lis[i]['scaf']) 
        lisofscaflen.append(len(lis[i]['scaf']))
    #This section will add into all the input values of scaffold into a lisofscaf. For staple strands, this line is optional
    
    for i in range(len(lis)):
        lisofstap.append(lis[i]['stap']) 
        lisofstaplen.append(len(lis[i]['stap']))
    #Similarly as above, the section will add the staple strands of the .json file into a lisofstap
    
    for i in range(len(lis)):
        lisofind.append(lis[i]['num'])
    #This section reads all 'num' section of the .json, creating a list ordered by when the numbers appear
    #in the .json indicating directionality of honey comb based on odd or even number
    
    
    for i in range(len(lis)):
        lisofstapind.append(lis[i]['num'])
    #This section is identical to the above function, just renamed as "lisofstapind"
    
    ### SUBSECTION 2: This section grabs general coordinate data and appends only the crossovers/square/arrows into a list ###
    
    for i in range(len(lis)):
        lisofstapL = []        #Serves to make the lisof___L/R lists empty again
        lisofstapR = []
        #First check if helice strand is odd or even for directionality
        if lisofstapind[i]% 2 == 0:   #Recall lisofstapind is a list of helice numbers, so we iterate over length of all strands
            directL = 2 #If our index is even number, left side of column index is 0
            directR = 0 #Right side of column index is 2
            directL_b = 3
            directR_b = 1 
        else:           #To summarize, this condition just switches which value in [a,b,c,d] we index depending on odd/even
            directL = 0 #If our index is odd, left side is at right position in index
            directR = 2 #right side is at left position in index
            directL_b = 1
            directR_b = 3
        for a in range(lisofstaplen[i]):
            #Now we check if crossover exists on left side of the helix
            if lisofstap[i][a][directL] != lisofstapind[i] and lisofstap[i][a][directL] != -1: #Conditions explained in report
                lisofstapL.append(lisofstap[i][a]) #Because crossover exists, append coordinates to a list
            elif lisofstap[i][a][directL] == -1 and lisofstap[i][a][directR] != -1: #Also appends boxes/arrows in left side
                lisofstapL.append(lisofstap[i][a])

            #Similarly check for crossover on right side of the helix
            if lisofstap[i][a][directR] != lisofstapind[i] and lisofstap[i][a][directR] != -1:
                lisofstapR.append(lisofstap[i][a]) #Similarly append these right side crossovers to a list
            elif lisofstap[i][a][directR] == -1 and lisofstap[i][a][directL] != -1: #Similarly appends right side box/arrows
                lisofstapR.append(lisofstap[i][a])

        lisofstapRnum.append(lisofstapR)    #These nested lists are the overall left/right side lists of DNA crossovers/arrow/
        lisofstapLnum.append(lisofstapL)    #square locations. Once the loop is finished, the original list becomes empty again

    ### SUBSECTION 3: This section has two functions, left and right, that iterate over a single DNA strand ###############
     ## The purpose of these essentially checks if a DNA strand has more than 1 crossover, iterating over a left direction ##
    ## and a rightward direction in the Cadnano DNA strand #################################################################
            
    # The input for Sandwichright is as follows (int, int, Sandwichsort, lisofstapRnum, lisofstapLnum, lisofstapind, int, int)
    
    # x is an int obtained from a for loop that looks for the length of the general list, meaning it tells us how many "helices"
    # there are overall
    
    # y is an int obtained from a for loop that looks for the length of the list from lisof_num. This is used to index
    # the location of a specific coordinate inside a single DNA helice. 
    
    # z is a list, Sandwichsort
    
    # r is a list, lisofstapRnum
    
    # l is a list, lisofstapLnum
    
    # i is a list, lisofstapind, which allows us to keep track of what helix number the script currently is looking at
    
    # d and de is the initialization condition that tells us whether or not we want to look at the pair [a,b] or [c,d] 
    # in the array
    
    def Sandwichright(x, y, z, r, l, i, d, de): # This searches crossovers for DNA moving left to right
        if r[x][y][d] != -1 and not r[x][y] in z:
            tmp = i.index(r[x][y][d]) #index number for specific helice i.e. where to go next
            for b in range(len(l[tmp])): # for how many crossovers/breaks are inside the left side list, iterate over this list
                if (r[x][y][d])%2 == 0: # If the helice number is an even number,
                    cal = 0
                    cbl = 1    #Switch the locations of where we index the array [a,b,c,d]
                    ccl = 2
                    cdl = 3
                    car = 2
                    cbr = 3
                    ccr = 0
                    cdr = 1
                else: #if the helice number is an odd number
                    cal = 2
                    cbl = 3
                    ccl = 0
                    cdl = 1
                    car = 0
                    cbr = 1
                    ccr = 2
                    cdr = 3
                if l[tmp][b][cdr] == r[x][y][de] + 1 and (r[x][y][d])%2 != 0: #This condition looks at the following.
                    z.append(r[x][y]) #If the next crossover (+1 from our original) is at a location that starts
                    z.append(l[tmp][b]) #From the left side crossover list (lisof__numL) then append these two values
                    Sandwichright(tmp, b, z, r, l, i, cal, cbl) #repeat this except at new variables at x = tmp, y = b, etc, essentially searching through "i" until the list is exhausted
                elif r[tmp][b][cdl] + 1 == r[x][y][de] and (r[x][y][d])%2 == 0: #This condition does the same as above
                    z.append(r[x][y]) #Except that if the next crossover is at a location that starts in (lisof_numR) then
                    z.append(r[tmp][b]) #We need to reverse directions, essentially calling sandwichleft instead.
                    Sandwichleft(tmp, b, z, r, l, i, car, cbr)
        elif r[x][y][d] == -1 and not r[x][y] in z: #This statement is essentially a termination condition i.e. arrow
            z.append(r[x][y])

    
    # The function for sandwichleft serves the same purpose as above, except the list that it indexes/uses in conditions
    # are now flipped from the lisof_numR to lisof_numL. All the inputs are the same as above.
    
    def Sandwichleft(x, y, z, r, l, i, d, de): #This searches crossovers for DNA moving right to left

        if l[x][y][d] != -1 and not l[x][y] in z:
            tmp = i.index(l[x][y][d]) #index number for specific helice i.e. where to go next
            for b in range(len(l[tmp])): # for how many crossovers/breaks are inside the left side list, iterate over this list
                if (l[x][y][d])%2 == 0: # If the helice number is an even number,
                    cal = 0
                    cbl = 1
                    ccl = 2
                    cdl = 3
                    car = 2
                    cbr = 3
                    ccr = 0
                    cdr = 1
                else: #if the helice number is an odd number
                    cal = 2
                    cbl = 3
                    ccl = 0
                    cdl = 1
                    car = 0
                    cbr = 1
                    ccr = 2
                    cdr = 3
                if r[tmp][b][cdl] +1 == l[x][y][de] and (l[x][y][d])%2 == 0: #if what we encounter in the right side list's crossover value
                    z.append(l[x][y])
                    z.append(r[tmp][b]) # equals the right side's crossover index value + 1... append that value
                    Sandwichleft(tmp, b, z, r, l, i, car, cbr) #repeat this except at new variables at x = tmp, y = b, etc.
                elif l[tmp][b][cdr] == l[x][y][de] +1 and (l[x][y][d])%2 != 0:
                    z.append(l[x][y])
                    z.append(l[tmp][b])
                    Sandwichright(tmp, b, z, r, l, i, cal, cbl)
        elif l[x][y][d] == -1 and not l[x][y] in z: #if left side list is not an arrow
            z.append(l[x][y])
        


######### SUBSECTION 4: Initialization of reading the list, i.e. determine if we start with a square, or a crossover    #######

#This condition below initializes for squares that move from the left side of the strand to the right in Cadnano

    t = 0
    at = 0
    for i in range(len(lis)): #Creates a list of DNA strands that start from left side square and end as right side arrow            
        for b in range(len(lisofstapLnum[i])):

            if lisofstapind[i]%2 != 0 and lisofstapLnum[i][b][0] == -1: #This condition is an assumptinon based on the helice number, more explained below
                Sandwichoverall.append([])
                Sandwichsort.append(lisofstapLnum[i][b])
                Sandwichright(i, b, Sandwichsort, lisofstapRnum, lisofstapLnum, lisofstapind, 2, 3)
                Sandwichoverall[at].extend(Sandwichsort[t:len(Sandwichsort)])                
                t = len(Sandwichsort)
                at += 1

# Once Sandwichright is completed, the list added to Sandwichsort should be re-organized specificly for a single DNA strand #
# This is then appended into Sandwichoverall.

    
    Sandwichsort = [] # We are clearing the initial list in Sandwich sort to "reset" what we're looking for

# The initialization condition below looks for squares that start on the right side and move to the left.
# We are assuming that movement from right to left means that a specific helice number must be even for a staple strand
# We also add a "checker" boolean that makes sure that we don't have any replicates inside our overall list, and if it
# Sees any replicates, we can skip that item and go onto the next. This is not so important in looking at squares, but
# Very important if we want to look at crossover initializations

    t = 0
    for i in range(len(lis)):
        for b in range(len(lisofstapRnum[i])):
            checker = lisofstapRnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            checker2 = lisofstapLnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            if lisofstapind[i]%2 == 0 and lisofstapRnum[i][b][0] == -1:   #From above, hence this condition here
                Sandwichoverall.append([])
                Sandwichsort.append(lisofstapRnum[i][b])
                Sandwichleft(i, b, Sandwichsort, lisofstapRnum, lisofstapLnum, lisofstapind, 2, 3)
                Sandwichoverall[at].extend(Sandwichsort[t:len(Sandwichsort)])
                t = len(Sandwichsort)
                at += 1

# Functionally, our initialization conditions are the same as above, except now, because there's no way to determine when
# we are "Terminating" our sequence, we add a check to make sure the next item in the list isn't already self contained in
# sandwichoverall. We do this twice for when we're initializing from the left side of the strand, or the right side of the
# strand.
                
    Sandwichsort = []
    t = 0
    for i in range(len(lis)):
        for b in range(len(lisofstapRnum[i])):
            checker = lisofstapRnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            checker2 = lisofstapLnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            if lisofstapind[i]%2 == 0 and lisofstapRnum[i][b][0] == -1 and not checker and not checker2:
                Sandwichoverall.append([])
                Sandwichsort.append(lisofstapRnum[i][b])
                Sandwichleft(i, b, Sandwichsort, lisofstapRnum, lisofstapLnum, lisofstapind, 2, 3)
                Sandwichoverall[at].extend(Sandwichsort[t:len(Sandwichsort)])
                t = len(Sandwichsort)
                at += 1
                
    Sandwichsort = []
    t = 0
    for i in range(len(lis)):
        for b in range(len(lisofstapRnum[i])):
            checker = lisofstapRnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            checker2 = lisofstapLnum[i][b] in (item for sublist in Sandwichoverall for item in sublist)
            if lisofstapind[i]%2 == 0 and not checker and not checker2:
                Sandwichoverall.append([])
                Sandwichsort.append(lisofstapRnum[i][b])
                Sandwichleft(i, b, Sandwichsort, lisofstapRnum, lisofstapLnum, lisofstapind, 2, 3)
                Sandwichoverall[at].extend(Sandwichsort[t:len(Sandwichsort)])
                t = len(Sandwichsort)
                at += 1

###############################################################################################################################
                
 # SECTION 1 END: This is the end of the section that checks for the location of crossovers in our .json

 # SECTION 2 START: This section works on the re-organized list in Sandwichoverall, and converts them into the following:
 
 # a list of lengths (int), as denoted by xstapcoord between two crossover/square/arrow points

 # a list of helice coordinates that correspond to xstapcoord, in order to keep track what length a specific DNA strand
 # is for a specific helice number. This means we can keep track for some helice "x" , it is "y" long.

 # And we use the same approach as before where we simply want to append xstapcoord and xstaphelicecoord into an overall list
 # since we're iterating over several DNA strands. This is represented by xstap_overall for a list of list of lengths
    
 # xloc_overall as well, is a list of list of helice coordinates.
    
    
    xstap_overall = []
    xloc_overall = []
    xstapcoord = []
    xstaphelicecoord = []
    for i in range(len(Sandwichoverall)): #We iterate the list for however long Sandwichoverall is
        if len(Sandwichoverall[i])%2 == 0 and int(len(Sandwichoverall[i])>2): #Checks if the length of the list inside
            z = 0 #is an even number and >2, which means it starts with a square and ends with an arrow.
            xloc = [] # The number of items must be bigger than 2 as well to make sure we have enough pairs to compare lengths.
            xstap_len = []
            for b in range(int(len(Sandwichoverall[i])/2)):
                if z>=len(Sandwichoverall[i]):
                    break
                elif Sandwichoverall[i][z][3]<Sandwichoverall[i][z+1][1]: #moving right
                    xstap_len.append(abs((Sandwichoverall[i][z][3] -1) - (Sandwichoverall[i][z+1][1] + 1)) +1)
                    xloc.append(Sandwichoverall[i][z][2])
                    xstapcoord.append([(Sandwichoverall[i][z][3] -1),(Sandwichoverall[i][z+1][1] + 1)])
                    xstaphelicecoord.append(Sandwichoverall[i][z][2])
                    z+=2
# This section essentially tries to give us values of an implicitly implied "third" coordinate. This is necessary to get the
# correct lengths when subtracting.
                elif Sandwichoverall[i][z][3]>Sandwichoverall[i][z+1][1]: #moving left
                    xstap_len.append(abs((Sandwichoverall[i][z][3] +1) - (Sandwichoverall[i][z+1][1] - 1)) +1)
                    xloc.append(Sandwichoverall[i][z][2])
                    xstapcoord.append([(Sandwichoverall[i][z][3] +1),(Sandwichoverall[i][z+1][1] - 1)])
                    xstaphelicecoord.append(Sandwichoverall[i][z][2])
                    z+=2
            if len(xloc) != 0:
                xstap_overall.append(xstap_len)
                xloc_overall.append(xloc)
        else:
            pass

# This looks at the similar as above, except now we are initializing with crossovers. This means we have n + 1 number of
# crossover locations in the strand. The number is kept odd, to also keep track of when these situations occur.

    for i in range(len(Sandwichoverall)):
        #xloc = []
        #xstap_len = []
        if len(Sandwichoverall[i])%2 != 0 and int(len(Sandwichoverall[i])) > 2: 
            z = 0 
            xloc = []
            xstap_len = []
            # Remove the last item on an odd list and perform our iterations.
            for b in range(int((len(Sandwichoverall[i])-1)/2)):
                if z>=len(Sandwichoverall[i]):
                    break
                elif Sandwichoverall[i][z][3]<Sandwichoverall[i][z+1][1]: #moving right
                    xstap_len.append(abs((Sandwichoverall[i][z][3] -1) - (Sandwichoverall[i][z+1][1] + 1)) +1)
                    xloc.append(Sandwichoverall[i][z][2])
                    xstapcoord.append([(Sandwichoverall[i][z][3] -1),(Sandwichoverall[i][z+1][1] + 1)])
                    xstaphelicecoord.append(Sandwichoverall[i][z][2])
                    z+=2
                elif Sandwichoverall[i][z][3]>Sandwichoverall[i][z+1][1]: #moving left
                    xstap_len.append(abs((Sandwichoverall[i][z][3] +1) - (Sandwichoverall[i][z+1][1] - 1)) +1)
                    xloc.append(Sandwichoverall[i][z][2])
                    xstapcoord.append([(Sandwichoverall[i][z][3] +1),(Sandwichoverall[i][z+1][1] - 1)])
                    xstaphelicecoord.append(Sandwichoverall[i][z][2])
                    z+=2
            if len(xloc) != 0:
                xstap_overall.append(xstap_len)
                xloc_overall.append(xloc)
        else:
            pass
    
    # Now, we are left with information regarding the following: A list of how long each strand is, with a list of helice
    # numbers that keep track exactly where that length of strand is. These lists are all nested in order to 
        
    if not xstap_overall:
        print("No sandwich strands detected in ", inputvar)
    else:
        print("Sandwich strands detected in ", inputvar)
        
        
        
    x = 0 # -> Just a counting variable
    d = 0 # -> Again, another counting variable
    type1 = 0
    type2 = 0
    for i in range(len(xstap_overall)):
        for b in range(len(xstap_overall[i])):
            if x < len(xstapcoord) or (b + 1) < len(xstap_overall[i]):
                if x >= (len(xstapcoord) - 1) or (b+1) >= len(xstap_overall[i]):
                    break
                elif xstap_overall[i][b] == xstap_overall[i][b+1]:
                    print("Between ", xstapcoord[x], "at helice ", xstaphelicecoord[x], "and ", xstapcoord[x+1], "at helice ", xstaphelicecoord[x+1], "exists a type 2 sandwich strand") #Sandwich type 2
                    x += 2
                    d += 1
                    type2 += 1
                elif xstap_overall[i][b] < xstap_overall[i][b+1]:
                    print("Between ", xstapcoord[x], "at helice ", xstaphelicecoord[x], "and ", xstapcoord[x+1], "at helice ", xstaphelicecoord[x+1], "exists a type 1 sandwich strand") #Sandwich type 1
                    x += 2
                    d += 1
                    type1 += 1
                else:
                    x += 2
                    
        
    print("The number of sandwich strand totals to be", str(d))
    print("There are", str(type1), "type 1 sandwich strands")
    print("There are", str(type2), "type 2 sandwich strands")
```
